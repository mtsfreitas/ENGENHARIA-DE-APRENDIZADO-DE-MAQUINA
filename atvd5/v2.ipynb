```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = pd.read_csv("breastcancerwisconsin.csv")
```

```python
# removendo colunas inúteis e mantendo apenas os atributos numéricos relevantes.
data = data.drop(['id', 'Unnamed: 32'], axis=1)

# mapeando valores númericos para o alvo "diagnosis"
# B (benigno) = 0 
# M (maligno) = 1

# Transformar a variável alvo (diagnosis) em numérica
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
```

```python
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

```python
# Padronizando os dados porque é uma etapa importante antes de aplicar o PCA (Principal Component Analysis) porque o PCA é sensível à escala das variáveis.
# A padronização transforma os dados para que cada atributo tenha média 0 e desvio padrão 1.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X.isnull().sum()
```

```python
# Aplicando PCA aos dados padronizados
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Dados originais
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Treinar e avaliar o modelo com dados originais
clf = LogisticRegression(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
acc_original = accuracy_score(y_test, y_pred)


# Dados transformados com PCA
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.3, random_state=42)

# Treinar e avaliar o modelo com dados transformados com PCA
clf_pca = LogisticRegression(random_state=42)
clf_pca.fit(X_train_pca, y_train_pca)
y_pred_pca = clf_pca.predict(X_test_pca)
acc_pca = accuracy_score(y_test_pca, y_pred_pca)
```

```python
print("Acurácia com dados originais:", acc_original)
print("Acurácia com dados transformados com PCA:", acc_pca)
```
